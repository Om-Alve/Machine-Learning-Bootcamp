```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('building.jpg',cv2.IMREAD_GRAYSCALE)
img = img / 255
```

```python
plt.imshow(img,cmap='gray')
print("Original Image")
```

```python
from scipy.signal import convolve2d
filterarr = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
convolved_image = convolve2d(img, filterarr, mode='same', boundary='wrap')
```

```python
plt.imshow(convolved_image,cmap='gray')
print("Horizontal Edges")
```

```python
filterarr = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
convolved_image = convolve2d(img, filterarr, mode='same', boundary='wrap')
plt.imshow(convolved_image,cmap='gray')
print("Vertical Edges")
```

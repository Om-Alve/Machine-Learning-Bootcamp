```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
X,y = spiral_data(samples=100,classes=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
class Layer:
    def __init__(self, n_inputs, n_neurons):
        # Initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
        
    def forwardprop(self,inputs):
        self.output = np.dot(inputs,self.weights) + self.biases
    
    def relu(self,inputs):
        
        self.output = np.maximum(0,inputs)
        
    def softmax(self,inputs):
        exp_vals = np.exp(inputs - np.max(inputs,keepdims=True))
        return exp_vals/np.sum(exp_vals,axis=1,keepdims=True)
```

```python
class CrossEntropyLoss:
    def calculate(self,ypred,ytrue):
        ypred = np.clip(ypred,1e-7,1-1e-7)
        return np.mean(-np.log(ypred[range(len(ypred)),ytrue]))
```

```python
layer = Layer(2,3)
layer.forwardprop(X)
```

```python
layer.relu(layer.output)
```

```python
layer2 = Layer(3,3)
layer2.forwardprop(layer.output)
out = layer2.softmax(layer2.output)
```

```python
cel = CrossEntropyLoss()
cel.calculate(out,y)
```

```python
#accuracy
acc = np.mean(np.argmax(out,axis=1)==y)

print("Accuracy : ",acc)
```

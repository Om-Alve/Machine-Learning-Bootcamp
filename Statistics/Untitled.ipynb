# Outlier removal

```python
df = sns.load_dataset('tips')
df.head()
data = np.sort(df['tip'])
```

```python
# Checking if it has any outliers
sns.boxplot(data)
```

```python
# defining lower bound and higher bound for outliers
Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)
IQR = Q3-Q1
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR
lower_fence,upper_fence
```

```python
clean_data = data[(data >= lower_fence) & (data <= upper_fence)]
```

```python
sns.boxplot(clean_data)
```

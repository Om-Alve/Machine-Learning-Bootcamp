```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.tan(x)
plt.plot(x,y)
```

```python
r = 5
theta = np.linspace(0,2*np.pi,100)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.figure(figsize=(4,4))
plt.plot(x,y)
```

```python
# linear data
x = np.linspace(2,20,1000)
m,c = 3,5
y = m*x+c
plt.scatter(x,y)
```

```python
noise = np.random.randn(1000)*5
y_mod = y+noise
plt.scatter(x,y_mod)
```

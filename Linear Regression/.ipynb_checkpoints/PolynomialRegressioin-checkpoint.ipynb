```python
class CustomLR:
    def __init__(self, lr=0.01,degree = 1):
        self.lr = lr
        self.degree = degree

    def polynomial_features(self,X):
        
        poly_X = np.ones((X.shape[0], 1))
        
        for i in range(1, self.degree + 1):
            feature = np.power(X,i)
            feature = feature.reshape(-1, 1) 
            poly_X = np.hstack([poly_X, feature])
            
        return poly_X
    
    def fit(self, X, y):
        mod_X = self.polynomial_features(X)
        self.theta_ = np.linalg.inv(mod_X.T @ mod_X) @ mod_X.T @ y

    def predict_point(self, X):
        X = self.polynomial_features(X)
        return np.dot(X, self.theta_).ravel()

    def predict(self, X):
        results = []
        for point in X:
            results.append(self.predict_point(point))
        return np.array(results)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
X = np.linspace(-2*np.pi,2*np.pi,50).reshape(-1,1)
y = np.cos(X)
```

```python
plt.scatter(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
plt.scatter(X_train,y_train)
```

```python
model = CustomLR(lr=0.01,degree=10)
model.fit(X_train,y_train)
```
